```python
f = open('purchase_log.txt')
import json
purchases = {}
i = 0
f.readline()
for line in f:
    # в каждом шаге цикла считывается ОДНА строка
    data = json.loads(line.strip())
    purchases[data['user_id']] = data['category']
        
    i += 1 
    if i > 100:
         break
        
for user_id, category in purchases.items():
    print(f"{user_id} ‘{category}‘")
```
